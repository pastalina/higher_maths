```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


def polar_to_decart(r, a):
  x = r * np.cos(a)
  y = r * np.sin(a)
  return x, y


print(polar_to_decart(2, np.pi/3))
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


a = np.arange(0, 2*np.pi, 0.01)
R = 2
r = 2*R*np.cos(a)
plt.polar(a, r)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.arange(np.pi/6, np.pi, 0.01)
plt.xlim(0, 2*np.pi)
y = 2*x + 3
plt.polar(x, y)
```
