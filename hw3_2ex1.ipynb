```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.arange(-6, 6, 0.05)
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.gca().set_aspect('equal', adjustable='box')
y1 = 2 * np.cos(x - 4) + 0
y2 = (-3) * np.cos(x - 0) + (-1)
y3 = 0.5 * np.cos(x - 3) + 5
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```
