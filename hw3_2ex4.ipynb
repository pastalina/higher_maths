```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.arange(-6, 10, 0.05)
plt.xlim(-6, 10)
plt.ylim(-2, 20)
# plt.gca().set_aspect('equal', adjustable='box')
y1 = x**2 - 1
y2 = 1 + (np.exp(x) - 1)/x
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

from scipy.optimize import fsolve
def equations(p):
  x, y = p
  return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

x1, y1 = fsolve(equations, (-2, 1))
print(x1, y1)
x2, y2 = fsolve(equations, (2, 5))
print(x2, y2)
x3, y3 = fsolve(equations, (4, 15))
print(x3, y3)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

from scipy.optimize import fsolve

def f(x):
  return np.exp(x) + x*(1 - x**2 + 1) - 1

x1 = fsolve(f, -2)
print(x1)
x2 = fsolve(f, 0)
print(x2)
x3 = fsolve(f, 3)
print(x3)
x4 = fsolve(f, 4)
print(x4)

x = np.arange(-6, 6, 0.05)
plt.xlim(-6, 6)
plt.ylim(0, 20)
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```
