```python
import random
import numpy as np
import matplotlib.pyplot as plt

n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, "o")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/((np.sum(x))**2 - n*np.sum(x**2))
b = (np.sum(y) - a*np.sum(x))/n
R = np.sum((x - np.sum(x)/n)*(y - np.sum(y)/n))/np.sqrt(np.sum((x - np.sum(x)/n)**2)*np.sum((y - np.sum(y)/n)**2))

A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]
c = np.corrcoef(x, y)

print(a, b)
print(a1, b1)
print(R)
print(c)
plt.plot([0, 1], [b, a + b])
plt.show()
```
