```python
import numpy as np
import itertools

i = 0
for p in itertools.product("ABCDEF", repeat=2):
  print("".join(p))
  i += 1
print("Число размещений 6 элементов по 2 с повторениями на практике:", i)
print("Число размещений 6 элементов по 2 с повторениями в теории:", np.power(6, 2))
```

```python
import math
import itertools

i = 0
for p in itertools.permutations("ABCDEF", 2):
  print("".join(p))
  i += 1
print("Число размещений 6 элементов по 2 без повторений на практике:", i)
print("Число размещений 6 элементов по 2 без повторений в теории:", math.factorial(6)/math.factorial(6 - 2))
```

```python
import math
import itertools

i = 0
for p in itertools.combinations("ABCDEF", 2):
  print("".join(p))
  i += 1
print("Число сочетаний 6 элементов по 2 на практике:", i)
print("Число сочетаний 6 элементов по 2 в теории:", math.factorial(6)/(math.factorial(2)*math.factorial(6 - 2)))
```

```python
import math
import itertools

i = 0
for p in itertools.permutations("ABCDEF", 6):
  print("".join(p))
  i += 1

print("Число перестановок 6 элементов на правктике:", i)
print("Число перестановок 6 элементов в теории:", math.factorial(6))
```
