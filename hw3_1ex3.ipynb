```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.arange(-20, 40, 0.0005)
plt.xlim(-30, 50)
plt.ylim(-35, 25)
plt.gca().set_aspect('equal', adjustable='box')
a = 10
b = -5
R = 25
y1 = np.sqrt(R**2 - (x-a)**2) + b
y2 = -np.sqrt(R**2 - (x-a)**2) + b
plt.plot(x, y1, color = "b")
plt.plot(x, y2, color = "b")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.arange(-10, 10, 0.0005)
plt.xlim(-10, 10)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal', adjustable='box')
a = 7
b = 4
y1 = np.sqrt(b**2 * (1 - x**2/a**2))
y2 = -np.sqrt(b**2 * (1 - x**2/a**2))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.arange(-30, 30, 0.0005)
plt.xlim(-30, 30)
plt.ylim(-20, 20)
plt.gca().set_aspect('equal', adjustable='box')
a = 7
b = 4
y1 = np.sqrt(b**2 * (1 + x**2/a**2))
y2 = -np.sqrt(b**2 * (1 + x**2/a**2))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```
