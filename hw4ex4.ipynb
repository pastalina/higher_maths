```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


a = np.arange(0.01, 0.02, 0.00001)
plt.xlim(0.01, 0.02)
plt.ylim(100, 500)
# plt.gca().set_aspect('equal', adjustable='box')
x1 = (np.arcsin(0) + np.pi)/a
x2 = (np.arcsin(0) + 2*np.pi)/a
x3 = (np.arcsin(0) + 3*np.pi)/a
plt.plot(a, x1)
plt.plot(a, x2)
plt.plot(a, x3)
plt.xlabel('a')
plt.ylabel('x')
plt.grid(True)
```
