```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
A1 = 2
B1 = 7
C1 = -4
Z1 = (A1*X +B1*Y)/(-C1) + 1
A2 = 3*A1
B2 = 3*B1
C2 = 3*C1
Z2 = (A2*X +B2*Y)/(-C2) - 10
ax.plot_wireframe(X, Y, Z1, color = "indigo")
ax.plot_surface(X, Y, Z2, color = "orange")
show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-7, 7, 0.01)
Y = np.arange(-7, 7, 0.01)
X, Y = np.meshgrid(X, Y)
A = 8
B = 6
C = 2
Z1 = np.sqrt((C**2) * (1 - (X**2)/(A**2) - (Y**2)/(B**2)))
Z2 = -np.sqrt((C**2) * (1 - (X**2)/(A**2) - (Y**2)/(B**2)))
Z3 = (X**2)/(A**2) - (Y**2)/(B**2)
ax.plot_wireframe(X, Y, Z1, color = "indigo")
ax.plot_wireframe(X, Y, Z2, color = "indigo")
ax.plot_surface(X, Y, Z3, color = "orange")
show()
```
